```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# import data
df = pd.read_csv('honeyproduction.csv')
df.head()
```

```python
# Look into data
prod_per_year = df.groupby('year').totalprod.mean().reset_index()

```

```python
# creating X axis from prod_per_year
X = prod_per_year.year
X = X.values.reshape(-1, 1)

# creating y axis from prod_per_year
y = prod_per_year.totalprod
```

```python
# plot the X and y axis
plt.scatter(X, y)
plt.show()
```

#### Creating and fitting a Linear Regression Model

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
regr.coef_, regr.intercept_
```

```python
# model predictions and plotting
y_predict = regr.predict(X)
plt.scatter(X, y)
plt.plot(X, y_predict)
plt.show()
```

#### Predicting the Honey Decline

```python
# create x values for the years 2013 - 2050
X_future = np.array(range(2013, 2051))
X_future = X_future.reshape(-1, 1)

# create y value predictions for the X_future values
future_predict = regr.predict(X_future)
```

```python
# plot the future predictions
plt.plot(X_future, future_predict)
plt.show()
```
